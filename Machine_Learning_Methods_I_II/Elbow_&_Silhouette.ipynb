```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

## Generate sample data (try to change)
X, _ = make_blobs(n_samples=500, centers=3, cluster_std=1.0, random_state=3)

## Plot the data
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title('Sample Data')
plt.show()
```

```python
## Define number of clusters to try
range_n_clusters = list(range(2, 8))

# Elbow method
inertia = []
for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=1)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 5))
plt.plot(range_n_clusters, inertia, marker='o')


## Look for the bend in the elbow (If there is no clear bend try the silhouette method)
plt.axvline(x=3, color='r', linestyle='--', label=f'Optimal Clusters: {3}')
plt.legend()
plt.title('Elbow Method For Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
## Silhouette method
silhouette_avg = []
for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=3)
    cluster_labels = kmeans.fit_predict(X)
    silhouette_avg.append(silhouette_score(X, cluster_labels))

plt.figure(figsize=(10, 5))
plt.plot(range_n_clusters, silhouette_avg, marker='o')
plt.axvline(x=3, color='r', linestyle='--', label=f'Optimal Clusters: {3}')
plt.legend()
plt.title('Silhouette Method For Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Average Silhouette Score')
plt.show()
```
